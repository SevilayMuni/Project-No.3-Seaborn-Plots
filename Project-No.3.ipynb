```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.impute import KNNImputer, SimpleImputer
```

```python
thyroid_df = pd.read_csv("Thyroid_Diff.csv", delimiter = ',')
thyroid_df
```

```python
# Renaming column names
thyroid_df = thyroid_df.rename(columns={'Hx Smoking': 'Smoking History',
                   'Hx Radiothreapy': 'Radiotherapy History',
                   'Pathology': 'Thyroid Cancer Type/Pathology',
                   'T': 'Tumor',
                   'N': 'Lymph Nodes',
                   'M': 'Metastasis',
                  'Response' : 'Response to Treatment'})
thyroid_df
```

```python
# Smoking History: whether individuals have a history of smoking (Yes or No).
# Radiotherapy History: whether individuals have radiotherapy treatment history (Yes or No).
# Thyroid Function: thyroid gland status or function.
# Physical Examination: Describes the results of a physical examination, likely related to the thyroid.
# Adenopathy: presence and location of adenopathy (enlarged lymph nodes).

# Thyroid Cancer Pathology: Describes the types of thyroid cancer based on pathology examinations, 
#including specific subtypes like "Micropapillary Papillary," "Follicular," and "HÃ¼rthle cell."

# Focality: whether the thyroid cancer is unifocal or multifocal.
# Risk: risk category associated with thyroid cancer.
# Tumor: stage of thyroid cancer, indicating the size and extent of the primary tumor.

# Lymph Nodes: indicating the involvement of nearby lymph nodes.
# Metastasis: whether the cancer has spread to distant organs.
# Stage: Represents the overall stage of thyroid cancer based on T, N, and M stage combination.

# Treatment Response: Describes treatment response 
# Categories; 'Indeterminate,' 'Excellent,' 'Structural Incomplete,' and 'Biochemical Incomplete.' 
# Recurred: Indicates whether thyroid cancer has recurred (e.g., Yes or No).
```

```python
# Replace values with clearer inputs
thyroid_df['Adenopathy'] = thyroid_df['Adenopathy'].replace({'No' : 'No Lympth Adenopathy' ,
                          'Left' : 'Left Side Body Adenopathy' ,
                          'Right' : 'Right Side Body Adenopathy' ,
                          'Extensive' : 'Extensive and Widespread Adenopathy'})
thyroid_df
```

```python
thyroid_df['Stage'] = thyroid_df['Stage'].replace({'I' : 'First-Stage' , 'II' : 'Second-Stage' , 'III' : 'Third-Stage'})

thyroid_df['Tumor'] = thyroid_df['Tumor'].replace({'T1a' : '1 cm or smaller', 
                             'T1b' : 'larger than 1 cm but smaller than 2 cm', 
                             'T2' : 'larger than 2 cm but smaller than 4 cm', 
                             'T3a' : 'larger than 4 cm',
                             'T3b' : ' grown outside the thyroid',
                             'T4a' : 'invaded nearby structures', 
                             'T4b' : 'invaded nearby structures'})

thyroid_df['Lymph Nodes'] = thyroid_df['Lymph Nodes'].replace({'N0' : 'no regional lymph node metastasis',
                                                               'N1b' : 'regional lymph node metastasis in the neck center',
                                                               'N1a' : 'regional lymph node metastasis in the neck lateral'})

thyroid_df['Metastasis'] = thyroid_df['Metastasis'].replace({'M0': 'no distant metastasis', 
                                                             'M1' : 'distant metastasis'})
thyroid_df
```

```python
thyroid_df.to_csv("Thyroid-Cancer-Preprocessed.csv", sep = ',', index = False)
```

```python
# Creating boxplot to explore the df

# 1x2 grid for the plots
fig, axes = plt.subplots(1, 2, figsize=(20, 8))

# Tumor Boxplot
sns.boxplot(x = 'Tumor', y = 'Age', data = thyroid_df, palette = "rocket_r", ax = axes[0])
axes[0].set_title('Age by Tumor Stage', fontsize = 16, fontweight = 'bold') 
axes[0].set_xlabel('Tumor Stage', fontsize = 14, fontweight = 'bold')
axes[0].set_ylabel('Age', fontsize = 14, fontweight = 'bold', labelpad = 20)
axes[0].tick_params(axis = 'x', rotation = 55)

# Stage Boxplot 
sns.boxplot(x = 'Stage', y = 'Age', data = thyroid_df, palette = "flare", ax = axes[1])
axes[1].set_title('Age by Cancer Stage', fontsize = 16, fontweight = 'bold')
axes[1].set_xlabel('Cancer Stage', fontsize = 14, fontweight = 'bold', labelpad = 20)
axes[1].set_ylabel('Age', fontsize = 14, fontweight = 'bold', labelpad = 20)
axes[1].tick_params(axis = 'x', rotation = 0)
sns.despine()
plt.savefig("Tumor-and-Cancer-Stage-Box-Plots.png", dpi = 300, bbox_inches = 'tight')
plt.show()
```

```python
def map_risk(risk):
    if risk == 'Low':
        return 1
    elif risk == 'Intermediate':
        return 2
    else:
        return 3

# New column with defined function
thyroid_df['Risk Level'] = thyroid_df['Risk'].apply(map_risk)
thyroid_df
```
